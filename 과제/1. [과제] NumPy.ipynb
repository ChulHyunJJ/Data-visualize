```python
import numpy as np
```

```python
# 모두 0으로 채운 3 x 3 배열 a 생성 
a = np.zeros((3,3))
print(a)

# 모두 1로 채운 2 x 3 배열 b 생성
b = np.ones((2,3))
print(b)

# 5로 채운 2 x 3 x 3 배열 c 생성
c = np.full((2,3,3),5)
print(c)

# 5 x 5 단위 행렬 d 생성
d = np.eye(5,5)
print(d)
```

```python
# range, reshape 함수를 사용하여 1~20 의 정수로 4 x 5 배열 arr 생성
arr = np.arange(1,21).reshape(4,5)
print(arr)
```

```python
# 불린 인덱스를 이용하여 arr 배열의 대각선 '\' 상의 요소 선택
bool1 = np.array([[True,False,False,False,False],
                  [False,True,False,False,False],
                  [False,False,True,False,False],
                  [False,False,False,True,False]])
print(arr[bool1])
```

```python
# arr 배열의 인덱스가 짝수인 요소만 조회 (불린 인덱스와 조건식을 조합)
bool2 = np.full((4,5),False)
print(bool2)

for i in range(0,4,2) :
  for j in range(0,5,2) :
    bool2[i,j] = True
  
print(bool2)

print(arr[bool2])
```

# 배열 간 연산

```python
m1 = np.array([
    [1,2],
    [3,4]
])
m2 = np.array([
    [5,6],
    [7,8]
])

# m1, m2 배열의 각 요소 간 합, 차, 곱 그리고 행렬곱을 구하기
print(np.add(m1,m2))
print(np.subtract(m1,m2))
print(np.multiply(m1,m2))
print(np.matmul(m1,m2))
```

```python
# arr 배열의 컬럼(열)별 합계
print(np.sum(arr,axis=0))
```

```python
# arr 배열의 행별 평균
print(np.mean(arr,axis=1))
```
