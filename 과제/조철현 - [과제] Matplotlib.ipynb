```python
# 데이터 읽어 오기. pandas의 read_csv 사용
data = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/data/tips.csv")
data
```

## 1. total_bill 과 tip 의 분포도(scatter) 시각화하기

### tip 금액에 따라 점의 크기 변화 주기, alpha 옵션으로 투명도 지정하기
### 축 label 추가하기, 플롯 title 추가하기


```python
x = data.total_bill
y = data.tip
size = y * 50
color = np.random.rand(244)

plt.scatter(x,y, s=size, c=color,marker='o', alpha=0.4, cmap='viridis')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.legend()
```

## 2. 성별에 따른 tip 비교하기 (bar plat)

```python
fig = plt.figure()
fig.subplots_adjust(hspace=1)
bar_width = 0.25
 
y1 = data.tip[(data.sex == 'Female')] 
y2 = data.tip[(data.sex == 'Male')]

x1 = np.arange(len(y1))
x2 = np.arange(len(y2))

plt.subplot(2,1,1)
plt.bar(x1,y1, color='royalblue', edgecolor='white', label='Female tip')
plt.ylabel("tip of Female")
plt.xlabel("people")
plt.title("tip comparison")
plt.subplot(2,1,2)
plt.bar(x2,y2, color='red', width=1, edgecolor='white', label='Male tip')
plt.ylabel("tip of Male")
plt.xlabel("people")
plt.title("tip comparison")
```

## 3. 남자의 여자의 팁 금액의 분포도(histogram)를 그려 비교하세요

### 2 X 1 의 subplot을 이용하세요.

```python
fig = plt.figure()
fig.subplots_adjust(hspace=1)
 
y1 = data.tip[(data.sex == 'Female')] 
y2 = data.tip[(data.sex == 'Male')]

plt.subplot(2,1,1)
plt.hist(y1, bins=50, alpha = 0.5, histtype = 'stepfilled', color='steelblue', edgecolor='none')
plt.ylabel("the number of people")
plt.xlabel("tip")
plt.title("tip comparison")
plt.subplot(2,1,2)
plt.hist(y2, bins=50, alpha = 0.5, histtype = 'stepfilled', color='skyblue', edgecolor='none')
plt.ylabel("the number of people")
plt.xlabel("tip")
plt.title("tip comparison")
```
